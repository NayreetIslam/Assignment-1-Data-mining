# Downloading Streaming data

We use the streaming API which collects all the upcoming tweets coming on the fly. We gather all the new tweets associated with jobs/ job postings using a wide variety of keywords. We can stop the downloading the streaming data by using kernel interrupt (in the Jupyter notebook) or by pressing Ctrl-C.

```python
from tweepy import Stream
from tweepy.streaming import StreamListener
 
class MyListener(StreamListener):
    def on_data(self, data):
        try:
            with open('jobpostings.json', 'a') as f:
                f.write(data)
                return True
        except BaseException as e:
            print("Error on_data: %s" % str(e))
        return True
 
    def on_error(self, status):
        print(status)
        return True
 
twitter_stream = Stream(auth, MyListener())
twitter_stream.filter(track=['job', 'hiring', 'apply', 'career', 'recruiting', 'vacancy notice', ' vacancy announcement', 'situation vacant', 'internship', 'salary', 'required qualification', 'skills', 'skillful', 'professional', 'qualified', 'ability to work', 'fulltime work', 'part time work', 'day work', 'we are in need of', 'we need', 'manager needed',  'director needed',  'employer needed', 'officer needed', 'developer needed', 'administrator needed', 'advisor needed', 'assistant needed', 'analyst needed', 'engineer needed', 'designer needed', 'technician needed', 'supervisor needed', 'consultant needed', 'physician needed', 'nurse needed', 'baby sitter needed', 'expert needed', 'manager wanted', 'director wanted', 'employer wanted', 'officer wanted', 'developer wanted', 'administrator wanted', 'advisor wanted', 'assistant wanted', 'analyst wanted', 'engineer wanted', 'designer wanted', 'technician wanted', 'supervisor wanted', 'consultant wanted', 'physician wanted', 'nurse wanted', 'baby sitter wanted', 'expert wanted', 'specialist needed', 'specialist wanted' ])
```

# Top 100 active users

We define a method called top_users() which takes the name of the file and the number of users as parameters and prints the name of the top active users in terms of number of tweets posted by them. The names of the top 100 active users is given below as the output.

```python
def top_users(tweetfile, k):
    user_list = {}
    fh = open(tweetfile,'r')
    for line in fh:
        try:
            tweet = json.loads(line)
        except:
            continue       
        if 'text' not in tweet:
            continue   
        u = tweet.get('user',{}).get('screen_name')
        user_list[u] = 1 + user_list.get(u, 0)
    users = list(user_list.items())
    sorted(users)
    for u in users[:k]:
            print(u)
```

```python
top_users('jobpostings.json', 100)
```

# Top 10 languages

Moreover, we also collect the information of top ten languages which was used to tweet the job related posts and show their information. 

```python
from collections import Counter
Counter = Counter(tweet_language)
# most_common() produces k frequently encountered
# input values and their respective counts.
most_occur = Counter.most_common(10)
print(most_occur)
key=[]
value=[]
for i in range(len(most_occur)):
    key.append(most_occur[i][0])
    value.append(most_occur[i][1])
    
print(key)
print(value)
```

# Top hashtages

```python
def top_hashtags(tweetfile, k):
    hashtag_list = {}
    fh = open(tweetfile,'r')
    # loop over lines in file
    for line in fh:
        # load json data
        try:
            tweet = json.loads(line)
        except:
            continue       
        if 'text' not in tweet:
            continue
        # extract hashtags entities    
        hts = tweet.get('entities',{}).get('hashtags')
        for hinfo in hts:
            h = hinfo['text']
            # add hashtag to list
            hashtag_list[h] = 1 + hashtag_list.get(h,0)
    # sort list of hashtags
    hts = list(hashtag_list.items())
    sorted(hts)
    # display top k hashtags
    for ht in hts[:k]:
            print(ht)
```

```python
top_hashtags('jobpostings.json',100)
```
